```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\priya\Downloads\students_data_2000_rows.csv")
df
```

```python
df2 = df.sort_values(by='Marks', ascending=False).sample(n=10)
df2
```

```python
plt.figure(figsize=(8,5))
plt.plot(df2['Name'], df2["Marks"], marker = '*', color = 'r')
plt.title("Marks of Students")
plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df2['Name'], df2["Marks"],)
plt.title("Marks of Students")
plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.barh(df2['Name'], df2["Marks"],)
plt.title("Marks of Students")
plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df2['Age'], df2["Marks"],)
plt.title("Marks of Students")
plt.xlabel("Age of Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df2["Marks"], bins = 12, color = 'blue', edgecolor = 'black')
plt.title("Marks of Students")
plt.xlabel("Name of Students")
plt.ylabel("Marks")
plt.show()
```

```python
dep = df["Department"].value_counts()
plt.figure(figsize=(8,5))
plt.pie(dep.values, labels=dep.index, autopct='%1.1f%%')
plt.title("Distribution of Students by Department")
plt.show()
```

```python
plt.figure(figsize=(6,6))

plt.plot(df2['Name'], df2['Marks'], marker = '*', color = 'r', label = 'Marks of Students')
plt.plot(df2['Name'], df2['Age'], marker = 'o', color = 'b', label = 'Age of Students')
plt.plot(df2['Name'], df2['Attendance'], marker = 's', color = 'g', label = 'Attendance of Students')
plt.title("Student Data")   
plt.xlabel("Name of Students")
plt.ylabel("Values")        
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,3,1) #[ 1 row, 3 columns, 1st subplot]
plt.bar(df2['Name'], df2["Marks"])
plt.title("Marks of Students")

plt.subplot(1,3,2)
plt.bar(df2['Name'], df2["Age"])
plt.title("Age of Students")

plt.subplot(1,3,3)
plt.plot(df2['Name'], df2["Attendance"]) 
plt.title("Attendance of Students")

plt.tight_layout()
plt.show()
```
