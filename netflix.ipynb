```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\priya\Downloads\netflix1.csv")
df
```

```python
df.duplicated().sum()
```

```python
df.head(10)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df,x = 'type')
plt.title('Count of Movies and TV Shows')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data = df, x = "rating",
                order = df["rating"].value_counts().index)
plt.title('Count of Ratings')
plt.show()
```

```python
plt.figure(figsize = (7,5))
sns.histplot(data = df,x = "release_year" , bins=30,kde=True)
plt.title("Distribution of Release Year")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df,x="type",y="release_year")
plt.title("Average release year by type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df,x="type",y="release_year")
plt.title("Average release year by type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.violinplot(data=df,x="type",y="release_year")
plt.title("Distribution of Release year by type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.stripplot(data=df,x="type",y="release_year")
plt.title("Distribution of Release year by type")
plt.show()
```

```python
Year_count = df.groupby('release_year').size().reset_index(name='count')
plt.figure(figsize=(10,5))
sns.lineplot(data=Year_count,x='release_year',y='count',marker='*')
plt.title('Number of Movies and TV Shows Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
sns.pairplot(df[['release_year','duration']])
plt.show()
```

```python
sns.jointplot(data=df,x='release_year',y='duration',kind='scatter')
plt.show()
```

```python
sns.jointplot(data=df.head(10),x='release_year',y='duration',kind='scatter')
plt.show()
```
