```python
import numpy as np
```

```python
score=np.random.randint(40,101,100)
print(score)
```

```python
mean_score = np.mean(score)
print(mean_score)
```

```python
median_score = np.median(score)
print(median_score)
```

```python
standard_deviation = np.std(score)
print(standard_deviation)
```

```python
above_average = score[score > mean_score]
count_above_average = len(above_average)
print(count_above_average)
print(above_average)
```

```python
marks = np.random.randint(0, 101, (5,3))
subjects = ['Math', 'Science', 'English']
print("Marks:")    
print(subjects)
print(marks)

total = np.sum(marks, axis=1)

percentage = total / 3

print("\nStudent\tTotal\tPercentage")
for i in range(5):
    print(f"{i+1}\t{total[i]}\t{percentage[i]:.2f}%")
```
